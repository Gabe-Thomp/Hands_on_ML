# Understanding Data for ML 

### Vocab

**Instance:** An instantiation of your data types (row)

**Feature:** A column of data

**Categorical Features vs Continuous Features:** <br>

Continous features are integer or float values. **Not all numeric values are continuous.** For example, a scale from 1-10.

**Dimensionality**: Number of features in a Dataset

**Sparsity**: The degree to which data exists in a dataset

### Visualize your Data

**Comparison Visualization**: Box plots good for categorical vs. continous

**Relationship Visualizations**: Scatterplots, Lineplots

**Distribution Visualization**: <br>*Histogram* uses bins for categorical data (Used to find mean)<br>
*Composition charts* Shows how much a subgroup contributes to total data

```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

vehicles = pd.read_csv('./Exercise Files/03/vehicles.csv')
vehicles.head()
```

## Relationship Visualization

```python
vehicles.plot(kind='scatter',x='citympg',y='co2emissions')
```

## Distribution Visualization

#### Use kind = 'hist' for histogram

```python
vehicles['co2emissions'].plot(kind = 'hist')
```

## Comparison Visualization

To create a box plot, must create a 'pivot table'

```python
vehicles.pivot(columns = 'drive',values = 'co2emissions')
```

```python
vehicles.pivot(columns = 'drive',values='co2emissions').plot(kind = 'box',figsize=[10,4])
```

## Composition Visualization

```python
year_group = vehicles.groupby('year')['drive']
year_group.value_counts().unstack()
```

Proportion of x wheel drive vehicles per year

```python
year_group.value_counts().unstack().plot(kind='bar',stacked=True,figsize=[10,6])
```

# Data Quality Issues

**Class imbalance issues**: Occur when the distribution of values for a class is not uniform. For example, a yes/no question where yes is often weighted more

### How to deal w/ missing data

```python
students = pd.read_excel('./Exercise Files/04/students.xlsx')
```

A mask object is a boolean array of same dimension as series

```python
mask = students['State'].isnull()
mask
```

Returns missing rows

```python
students[mask]
```

### Remove missing values

Removes all rows with **any** missing values

```python
students.dropna()
```

#### Drops columns w/ missing values

```python
students.dropna(axis=1)
```

#### Targets certain Rows

```python
students.dropna(subset=['State','Zip'],how='all')
```

#### Targets certain columns

thresh=10: requires 10 non-na values

```python
students.dropna(axis=1,thresh=10)
```

### Resolve Missing Values

#### Fills with preset value

```python
students.fillna({'Gender':'Female'}).head()
```

#### Fills with median value

```python
students.fillna({'Age':students['Age'].median()}).head()
```

Filtering for Granger, Indiana

```python
mask = (students['City']=='Granger') & (students['State']=='IN')
```

```python
students.loc[mask,:]
```

```python
students.loc[mask,'Zip'] = 46530
students
```
