# Support Vector Machines 

### What is it?

A *Support Vector Machine* is a model capable of linear or nonlinear classification, regression, or outlier detection. Well suited for complex, small or medium sized datasets.

## Linear SVM Classification

Lines can be used to separate classes on a graph. If this line is straight, they are *linearly separable* classes. Essentially, an SVM classifer fits the "widest possible street" between classes, known as *large margin classification*. By doing this, it is good at generalizing to new data as well at fitting current data.

This means that it is highly dependent on the instances at the "fringes" between classes. These instances are called the *support vectors*. 

#### Note: Scaling is important

## Soft Margin Classificaiton

***Hard margin classification***: Essentialy, restrictions are strictly imposed on which instances can be on which side of a line
- Only works w/ linearly separable data
- Very sensitive to outliers

For example, if there are two separate classes and somehow one ends up in another, it is impossible to separate them.

More flexible models are avaiable - a balance between a wide road and limiting *margin violations* (the instances on either side of the street)

#### Hyperparameters

**C**: A hyperparameter that controls the balance between margin violations and a skinny road. A lower C means more margin violations.

**Note**: If model is overfitting, it can be regularized by reducing C

Note also: **hinge** loss hyperparameter

```python
import numpy as np
from sklearn import datasets
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
from matplotlib import pyplot as plt
%matplotlib inline

iris = datasets.load_iris()
X = iris["data"][:, (2,3)] # petal length, petal width
y = (iris["target"] == 2).astype(np.float64) # Iris virginica

svm_clf = Pipeline([
            ("scaler", StandardScaler()),
            ("linear_svc", LinearSVC(C=1, loss="hinge")),
        ])

svm_clf.fit(X, y)
```

```python
svm_clf.predict([[5.5, 1.6]])
```

Note: SVM do not output probabilities! Maybe because there is no "score" function associated with the point?

## Nonlinear SVM Classification

Not all datasets are linearly separable! However, there are methods for dealing with this. For example, polynomial features can be added that are linearly separable. 

To implement this idea, create a Pipeline with `PolynomialFeatures` transformer, a `StandardScaler` and a `LinearSVC`.

```python
from sklearn.preprocessing import PolynomialFeatures

polyfeat = PolynomialFeatures()
scaler = StandardScaler()
svc = LinearSVC(loss = "hinge", C=10)
polynomial_svm_clf = Pipeline([
    ("polynomial", polyfeat),
    ("scaler", scaler),
    ("svc", svc)
])

polynomial_svm_clf.fit(X, y)
```

### Polynomial Kernel

Polynomial features is great, but there are drawbacks
- At a low degree, it is not good for complex datasets
- At a high degree, it creates a *lot* of features and is slower

With SVMs, you can apply a trick called the *kernel trick*, which originates from mathematics. It gets rid of combinatorial explosion

Trains an svm classifier with a third degree polynomial kernel. If model is 
- *Underfitting*, may increase polynomial degree
- *Overfitting*, may decrease polynomial degree

```python
from sklearn.svm import SVC
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=100, noise=0.15)

poly_kernel_svm_clf = Pipeline([
    ('scaler', StandardScaler()),
    ('svm_clf', SVC(kernel="poly", degree=3, coef0=1, C=5))
])

poly_kernel_svm_clf.fit(X,y)
```

## Similarity Features

The use of a *similarity function* is used to determine how far an instance is from a *landmark*.

For example, the Gaussian *Radial Basis Function (RBF)* <br>
$\phi_\gamma(\bf{x},\ell) = exp(-\gamma||\bf{x}-\ell||^2)$
As $|\gamma|$ increases, the curve gets tighter. As a result, each instance's range of influence is smaller. The decision boundary is more irregular. 

$\gamma$ is like a regularization hyperparameter. If it is too high, it ends up getting around single instances. Too low, and the data is *underfit*

This function is centered at the landmark and output a value from 0 to 1 based on proximity to landmark

Why do we use these? **They often make linearly separable variables**

The simplest strategy is to create a landmark at *every instance* in the dataset. This converts an $m\times n$ training set to an $m\times m$ set.

## Gaussian RBF Kernel

Again, this kernel works to simulate an RBF without creating all of the additional features

```python
rbf_kernel_svm_clf = Pipeline([
    ("scaler", StandardScaler()),
    ("svm_clf", SVC(kernel = "rbf", gamma=5, C=0.001))
])

rbf_kernel_svm_clf.fit(X, y)
```

### Note:

Other kernels exist but are not used as much. There are some for specific data structures like text and DNA.

### Which one do I use?

- Try linear first, where `LinearSVC` is much faster than `SVC(kernel=linear)`
- Cross Validation and grid search is also very important for finding good hyperparameters

## SVM Regression

Here, the *smallest possible street* is fitted to the data, while minimizing the **margin violations** (outliers). The width of the street is controlled by **$\epsilon$** . Adding more training instances within the margin would not affect the predictions, therefore it is said to be *$\epsilon$-insensitive.*

This is the sklearn package `from sklearn.svm import LinearSVR`

sklearn's `SVR` class can also be used, which supports the *kernel trick*. Here,

`from sklearn.svm import SVR`<br>
`svm_poly_reg = SVR(kernel="poly", degree=2, C=100, epsilon=0.1)`

Pro: the LinearSVR scales linearly w/ the size of the dataset, which is really nice

```python
plt.scatter(X[:,0], X[:,1])
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
svc_clf = Pipeline([
    ("scaler", StandardScaler()),
    ("clf", SVC(kernel='linear'))
])

scores = []
for _ in range(1000):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2)

    svc_clf.fit(X_train, y_train)

    y_preds = svc_clf.predict(X_test)
    score = (y_preds==y_test).mean()
    scores.append(score)
sum(scores)/len(scores)
```

Score centered around ~95

Note: with a low C (high regularization, underfit), the score is a *lot* worse

```python
svc_clf = Pipeline([
    ("scaler", StandardScaler()),
    ("clf", SVC(kernel='linear', C=.01))
])

scores = []
for _ in range(1000):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2)

    svc_clf.fit(X_train, y_train)

    y_preds = svc_clf.predict(X_test)
    score = (y_preds==y_test).mean()
    scores.append(score)
sum(scores)/len(scores)
```

It performs poorly on training data, means that it is likely *underfit*

```python
(svc_clf.predict(X_train) == y_train).mean()
```

#### SGD Classifier

```python
from sklearn.linear_model import SGDClassifier

svc_clf = Pipeline([
    ("scaler", StandardScaler()),
    ("clf", SGDClassifier())
])

scores = []
for _ in range(1000):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2)

    svc_clf.fit(X_train, y_train)

    y_preds = svc_clf.predict(X_test)
    score = (y_preds==y_test).mean()
    scores.append(score)
sum(scores)/len(scores)
```

Similar results above!
