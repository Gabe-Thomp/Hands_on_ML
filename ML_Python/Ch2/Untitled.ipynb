```python
import matplotlib.pyplot as plt
import math
import numpy as np
import matplotlib

%matplotlib inline
```

```python
t = np.linspace(-2*math.pi,2*math.pi)

A_x = 1 
A_y = 1
omega = 1

phi = np.linspace(0,2*np.pi)
num = len(phi)
numt = len(t)


fig, ax = plt.subplots()


x = A_x*np.cos(omega*t)
y = A_y*np.cos(omega*t+phi[0])
sc = ax.scatter(x, y, alpha=.5)

def animate(i):
    x = A_x*np.cos(omega*t)
    y = A_y*np.cos(omega*t+phi[i])
    
    sc.set_offsets(np.c_[x,y])
    return sc
    
ani = matplotlib.animation.FuncAnimation(fig, animate, 
                frames=50, interval=300, repeat=True) 

plt.show()
ani.save(".mp4")
```

```python
for i in range(len(phi)):
    x = A_x*np.cos(omega*t)
    y = A_y*np.cos(omega*t+phi[i])

    plt.scatter(x,y,alpha=.5)
    plt.show()
```
