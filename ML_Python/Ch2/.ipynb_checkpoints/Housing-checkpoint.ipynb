## Importing Data from GitHub

```python
import tarfile
import os
import urllib
import numpy as np
```

```python
DOWNLOAD_ROOT = "https://raw.githubusercontent.com/ageron/handson-ml2/master/"
HOUSING_PATH = os.path.join("datasets", "housing")
HOUSING_URL = DOWNLOAD_ROOT + "datasets/housing/housing.tgz"
def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH): 
    os.makedirs(housing_path, exist_ok=True)
    tgz_path = os.path.join(housing_path, "housing.tgz") 
    urllib.request.urlretrieve(housing_url, tgz_path)
    housing_tgz = tarfile.open(tgz_path)
    housing_tgz.extractall(path=housing_path)
    housing_tgz.close()
```

## Creating Test Set

Pick random 20% of data and set it aside. You may want to pick a random 20% at the beginning and set it aside. This can be done with a **numpy seed** or by saving the data at the beginning.

**Both of these implementations will break** the next time you fetch an updated dataset, however.

Instead, an instance's unique *identifier* can be used to decide whether or not it goes into test set. Compute a hash of each identifier and put it in test set if hash is $\leq20\%$ of the maximum hash value.

**Code Wiki**: A hash function is any function that can be used to map data of arbitrary size to fixed-size values. The values returned by a hash function are called hash values, hash codes, digests, or simply hashes. The values are used to index a fixed-size table called a hash table. Use of a hash function to index a hash table is called hashing or scatter storage addressing

**zlib:** lossless data-compression library for use on virtually any computer hardware and operating system. <br><br>
**crc32**: CRC32 is an error-detecting function that uses a CRC32 algorithm to detect changes between source and target data. The CRC32 function converts a variable-length string into an 8-character string that is a text representation of the hexadecimal value of a 32 bit-binary sequence.

```python
from zlib import crc32
def test_set_check(identifier, test_ratio):
    return crc32(np.int64(identifier)) & 0xffffffff < test_ratio * 2**32
def split_train_test_by_id(data, test_ratio, id_column):
    ids = data[id_column]
    in_test_set = ids.apply(lambda id_: test_set_check(id_, test_ratio)) 
    return data.loc[~in_test_set], data.loc[in_test_set]
```

To solve this problem, **one-hot encoding** is often used. <br>
**one-hot encoding** is where one binary attribute per category is created, which is equal to 1 when an instance is a member of the category. 

**one-hot** because *one* attribute is active, or *hot*, while the others are zero, prd *cold*.

The new attributes are sometimes called **dummy attributes**

## Custom Transformers

Using sklearn **TransformerMixin** as a base class gets you fit_transform method. The **BaseEstimator** base class will activate two more methods: **get_params()** and **set_params()**, which will allow for tuning hyperparameters

```python
from sklearn.base import BaseEstimator, TransformerMixin
rooms_ix, bedrooms_ix, population_ix, households_ix = 3, 4, 5, 6

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room=True): # no *args or **kargs
        self.add_bedrooms_per_room = add_bedrooms_per_room 
    
    def fit(self, X, y=None):
        return self # nothing else to do 

    def transform(self, X):
        rooms_per_household = X[:, rooms_ix] / X[:, households_ix] 
        population_per_household = X[:, population_ix] / X[:, households_ix] 
    
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
            return np.c_[X, rooms_per_household, population_per_household,
                       bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_household, population_per_household]

attr_adder = CombinedAttributesAdder(add_bedrooms_per_room=False)
housing_extra_attribs = attr_adder.transform(housing.values)
```

## Feature Scaling

Machine Learning algorithms don't perform well when input attributes have different scales. For example, housing data ranges [6, 39000], while median income ranges from [0,15].

Two methods of scaling: **min-max scaling** and **standardization**. 
<br>

**Minmax scaling**, also known as normalization, shifts values to range from 0 to 1. Sklearn has method *MinMaxScaler* for this.
<br>

**Standardization** subtracts mean value (so that standardized values have zero mean) and divides by standard deviation (so that the distribution has standard variance). Though it doesn't have a set range, it is much less affected by outliers. Sklearn *StandardScaler* is good for this.

## Transformation Pipelines

A pipeline is a series of functions/methods that data is run through.

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

num_pipeline = Pipeline([
    ('imputer',SimpleImputer(strategy='median')),
    ('atttribs_adder', CombinedAttributesAdder()),
    ('std_scaler', StandardScaler())
])

housing_num_tr = num_pipeline.fit_transform(housing_num)
```

Pipeline constructor takes a list of name/estimator pairs defining a sequence of steps. All but the last estimator must be transformers.

### Column Transformer

It would be nice to have a transformer that works for columns, though. Then, we don't have to split up numerical and categorical data.

```python
from sklearn.compose import ColumnTransformer

num_attribs = list(housing_num)
cat_attribs = ["ocean_proximity"]

full_pipeline = ColumnTransformer([
    ("num", num_pipeline, num_attribs),
    ("cat", OneHotEncoder(), cat_attribs),
])


housing_prepared = full_pipeline.fit_transform(housing)
```

We use sklearn **ColumnTransformer** class. Requires *list of tuples* with [(name, transformer, list of column names/indices)]
